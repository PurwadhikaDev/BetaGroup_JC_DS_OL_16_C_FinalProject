# **1. Business Problem Understanding**
<hr>

**Context**  


Dalam industri perhotelan yang dinamis dan kompetitif, **permintaan pemesanan kamar** merupakan **faktor kunci determinan** bagi keberhasilan operasional. Kemampuan memahami fluktuasi dan pola permintaan (*demand forecasting*) menjadi **aspek kritis** untuk **optimalisasi pendapatan (*revenue optimization*)** dan **stabilitas tingkat okupansi (*occupancy rate*)**.  Di tengah pengaruh variabel musiman, tren perjalanan, dan dinamika ekonomi global, **prediksi permintaan (*predictive analytics*)** dan kemampuan merespons secara dinamis adalah **keunggulan kompetitif strategis**. Pendekatan berbasis data memungkinkan hotel memaksimalkan potensi pendapatan sekaligus memitigasi risiko kerugian akibat kamar kosong (*unsold inventory*).

**Kontrak pemesanan di muka (*advance booking*)** menjadi komponen penting dalam ekosistem ini. Meskipun memberikan proyeksi kinerja awal, **risiko pembatalan (*cancelation rate*)** dan ketidakhadiran tamu (*no-show*) menghadirkan tantangan dalam manajemen inventaris. Hotel menghadapi dilema: menjamin ketersediaan kamar untuk pemesanan terkonfirmasi versus potensi *lost revenue* akibat kamar kosong. Pemanfaatan **analisis data historis** – termasuk pola pembatalan, musiman, dan perilaku konsumen –  penting untuk membangun model prediktif, mengoptimalkan kebijakan *overbooking*, dan memitigasi risiko.

**Pentingnya Prediksi Pembatalan, Terutama Saat *Peak Season***

Prediksi pembatalan pemesanan menjadi semakin krusial, terutama saat *peak season*. Langkah pencegahan terhadap pelanggan yang berpotensi membatalkan pesanan sangat vital karena kerugian akibat pembatalan di periode ramai dapat berlipat ganda. Hotel dengan **tingkat okupansi tinggi** namun **tingkat pembatalan tinggi** justru dapat lebih merugi dibandingkan hotel dengan tingkat okupansi dan pembatalan yang sama-sama rendah.

**Kerugian Akibat Tingkat Pembatalan Tinggi pada Hotel dengan Okupansi Tinggi:**

1.  **Kerugian Pendapatan Signifikan:** Pembatalan saat okupansi tinggi menghilangkan potensi pendapatan maksimal dari kamar yang seharusnya terisi penuh.
2.  **Biaya Operasional Tetap:** Biaya persiapan kamar dan staf tetap harus ditanggung, meskipun tamu batal hadir, memperbesar kerugian relatif.
3.  **Hilangnya Peluang Pendapatan Tambahan:** Kamar yang dibatalkan, terutama menjelang tanggal menginap, menghilangkan kesempatan untuk dijual kembali pada permintaan tinggi.
4.  **Gangguan Operasional dan Efisiensi:** Pembatalan massal mengacaukan manajemen inventaris dan perencanaan, menurunkan efisiensi, dan meningkatkan biaya operasional.

**Kesimpulan:**

Hotel perlu proaktif mengelola pembatalan dan meningkatkan strategi manajemen untuk meminimalkan kerugian akibat tingkat pembatalan yang tinggi. Prediksi pembatalan pemesanan yang akurat menjadi solusi krusial untuk optimalisasi operasional dan pendapatan hotel.

## **1.1 Problem Statement**

**Tantangan Prediksi Pembatalan Pemesanan Hotel:**

Akurasi prediksi pembatalan pemesanan hotel merupakan isu krusial yang berdampak signifikan pada kinerja operasional dan finansial. Ketidakmampuan memprediksi pembatalan secara tepat dapat memicu dua risiko utama:

*   **Overbooking:**  Antisipasi pembatalan berlebihan menyebabkan kamar tidak tersedia saat tamu datang, menurunkan kepuasan dan merusak reputasi hotel. Kompensasi mungkin diperlukan.
*   **Underbooking:** Kurangnya prediksi pembatalan mengakibatkan banyak kamar kosong tidak terjual, mengurangi pendapatan dan efisiensi aset.

**Dampak Negatif Prediksi Pembatalan yang Tidak Akurat:**

Ketidakakuratan prediksi pembatalan pemesanan merugikan hotel dalam berbagai aspek, meliputi:

*   **Operasional:** Kesulitan mengatur kapasitas kamar dan mengelola sumber daya secara efisien.
*   **Finansial:**  Penentuan strategi harga yang tidak optimal dan potensi kerugian pendapatan.
*   **Kepuasan Pelanggan:** Potensi penurunan kepuasan akibat *overbooking*.

**Kebutuhan Solusi Prediktif:**

Oleh karena itu, diperlukan solusi prediktif yang akurat untuk membantu hotel mengantisipasi pembatalan. Solusi ini akan memungkinkan hotel untuk mengambil tindakan proaktif dalam:

*   **Mengurangi Risiko:** Meminimalkan potensi *overbooking* dan *underbooking*.
*   **Meningkatkan Kinerja:**  Mengoptimalkan pendapatan dan efisiensi operasional secara keseluruhan.

    *   Reduksi risiko *overbooking* dan *underbooking*.
    *   Peningkatan efisiensi operasional dan pendapatan hotel.
    *   Peningkatan kepuasan tamu melalui manajemen pemesanan yang lebih baik.
    *   Pengambilan keputusan bisnis yang lebih tepat berbasis data dan prediksi.

- **True Positive (TP):** Pelanggan diprediksi akan membatalkan dan benar-benar membatalkan.
- **False Positive (FP):** Pelanggan diprediksi akan membatalkan, tetapi sebenarnya tidak membatalkan.
- **False Negative (FN):** Pelanggan diprediksi tidak akan membatalkan, tetapi sebenarnya membatalkan.
- **True Negative (TN):** Pelanggan diprediksi tidak akan membatalkan dan benar-benar tidak membatalkan.



### **Evaluasi Kinerja Model Prediksi Pembatalan Pemesanan Hotel**

Untuk mengukur efektivitas model dalam memprediksi pembatalan pemesanan, berbagai metrik evaluasi akan digunakan untuk menilai tingkat kesalahan model.  Fokus utama evaluasi ini adalah **optimalisasi pendapatan** dengan memprioritaskan **akurasi model dalam mengidentifikasi pemesanan yang akan dibatalkan (kelas positif)** dan **meminimalisir kerugian akibat kesalahan prediksi** (baik *False Negative* maupun *False Positive*).

Dalam konteks ini, target prediksi model adalah:

*   **Target:**
    *   **0:** Pemesanan **Tidak Dibatalkan**
    *   **1:** Pemesanan **Dibatalkan**

**Tindakan Hotel Berdasarkan Prediksi Model:**

Berdasarkan prediksi pembatalan, hotel dapat mengambil tindakan strategis seperti:

*   **Kebijakan Mitigasi Pembatalan:**  Implementasi *double booking* untuk pemesanan yang diprediksi akan dibatalkan atau penerapan kebijakan deposit untuk pemesanan baru. 

**Catatan:** Kebijakan deposit perlu dipertimbangkan dengan cermat karena deposit yang terlalu tinggi berpotensi mengurangi daya tarik hotel bagi calon pelanggan.

**Konsekuensi Kesalahan Prediksi:**

Evaluasi kinerja model akan fokus pada dua jenis kesalahan utama dan dampaknya bagi hotel:

#### **Kesalahan Tipe I: *False Positive***

*   **Definisi:** Model memprediksi pemesanan **akan dibatalkan**, padahal sebenarnya **tidak dibatalkan**.
*   **Konsekuensi Finansial:**  Penerapan promosi atau kebijakan mitigasi (seperti deposit) pada pelanggan yang sebenarnya tidak akan membatalkan menjadi **inefisien**. Hotel berpotensi kehilangan **optimalisasi pendapatan** karena mengeluarkan biaya pemasaran (estimasi **11% dari harga kamar malam**) yang tidak perlu. Biaya ini mencakup estimasi **4-5% untuk pemasaran digital** dan **6% untuk pemasaran konvensional**, serta potensi biaya tambahan untuk **menangani dampak ulasan negatif akibat *double booking***.

#### **Kesalahan Tipe II: *False Negative***

*   **Definisi:** Model memprediksi pemesanan **tidak akan dibatalkan**, padahal sebenarnya **dibatalkan**.
*   **Konsekuensi Finansial:** Hotel **kehilangan pendapatan** secara langsung sebesar **harga pemesanan per malam** untuk kamar yang batal terisi.

**Estimasi Dampak Biaya (Kuantifikasi Konsekuensi):**

Untuk mengukur dampak finansial dari kesalahan prediksi, digunakan asumsi biaya berdasarkan data industri perhotelan di Portugal:

*   **Harga Rata-rata Kamar Hotel per Malam di Portugal:** 91.85 Euro
*   **Estimasi Biaya Pemasaran (termasuk gaji tim pemasaran dan dampak ulasan negatif):** **11% dari harga kamar per malam = ±10 Euro**

Tentu, berikut adalah penataan ulang struktur bahasa, perbaikan tata bahasa, dan format yang bersih untuk teks Anda yang menjelaskan pemilihan F2-Score sebagai metrik evaluasi utama, agar mudah di-*copy-paste*:

#### **Metrik Evaluasi Utama: F2-Score – Memprioritaskan Minimasi *False Negative* untuk Optimalisasi Keuntungan Hotel**

Dalam konteks operasional bisnis perhotelan, baik kesalahan prediksi *False Positive* (prediksi positif palsu) maupun *False Negative* (prediksi negatif palsu) **keduanya berpotensi menimbulkan kerugian finansial bagi hotel**.  Namun demikian, perlu digarisbawahi bahwa **dampak kerugian yang diakibatkan oleh *False Negative* cenderung lebih signifikan dibandingkan dengan *False Positive***.

Oleh karena pertimbangan tersebut, **F2-Score ditetapkan sebagai metrik evaluasi utama dalam pemodelan ini**.  Pemilihan F2-Score didasarkan pada karakteristiknya yang **memberikan penekanan lebih besar pada *Recall***.  Fokus pada peningkatan *Recall* sangat relevan dalam konteks bisnis hotel karena memungkinkan model untuk dioptimalkan dengan tujuan utama: **meminimalisir kesalahan *False Negative*** (yaitu, kesalahan memprediksi bahwa pemesanan tidak akan dibatalkan, padahal kenyataannya dibatalkan) dan pada akhirnya **memaksimalkan potensi keuntungan hotel**.

**Alasan Pemilihan F2-Score:**

*   **Kerugian Akibat *False Negative* Lebih Besar**: Dalam skenario *False Negative*, model gagal mendeteksi pemesanan yang sebenarnya akan dibatalkan. Akibatnya, hotel **kehilangan peluang untuk melakukan *overbooking* atau mengambil tindakan proaktif lainnya untuk mengisi kamar yang kosong tersebut**. Kamar tetap kosong tanpa terisi, yang berarti **potensi pendapatan dari kamar tersebut hilang**.

*   **Kerugian Akibat *False Positive* Lebih Terkelola**: Sebaliknya, dalam kasus *False Positive*, model secara keliru memprediksi bahwa pemesanan akan dibatalkan, padahal sebenarnya tidak.  Meskipun ini dapat menyebabkan hotel mengambil tindakan yang tidak perlu (seperti menawarkan kompensasi karena *overbooking* yang tidak seharusnya terjadi), **kerugian finansial akibat *False Positive* umumnya lebih kecil dan lebih terkelola** dibandingkan dengan kehilangan pendapatan akibat kamar kosong (*False Negative*). Hotel masih memiliki potensi untuk mendapatkan pendapatan dari pemesanan awal yang sebenarnya tidak dibatalkan.

*   **Fokus F2-Score pada *Recall***: F2-Score adalah varian dari F-score yang secara khusus **memberikan bobot lebih besar pada *Recall* dibandingkan dengan *Precision***.  Dalam F2-Score, *Recall* memiliki bobot dua kali lipat dibandingkan *Precision*.  Karakteristik ini **sangat sesuai dengan tujuan bisnis hotel** yang lebih mengutamakan **minimasi *False Negative*** (memastikan sebanyak mungkin pemesanan yang berpotensi batal terdeteksi) daripada memaksimalkan *Precision* (memastikan bahwa setiap pemesanan yang diprediksi batal benar-benar batal).

Dengan menjadikan **F2-Score sebagai metrik evaluasi utama**, pemodelan ini diarahkan untuk menghasilkan model *machine learning* yang **sangat sensitif dalam mengidentifikasi potensi pembatalan reservasi**.  Prioritas utama adalah **meminimalisir *False Negative***, yang secara langsung berkontribusi pada **pengoptimalan tingkat hunian kamar dan peningkatan pendapatan hotel secara keseluruhan**.  Meskipun *False Positive* tetap menjadi perhatian, dampaknya dianggap lebih kecil dan dapat dikelola dibandingkan dengan kerugian signifikan yang timbul akibat *False Negative* dalam konteks bisnis perhotelan.

**Tahap *Data Understanding* dalam Analisis Data**

**Data Cleaning (Pembersihan Data): Definisi dan Pentingnya**

**Mengapa Data Cleaning Penting?**

Data yang **bersih dan akurat** krusial untuk **validitas dan keandalan hasil analisis**. Data yang "kotor" dapat menimbulkan dampak negatif:

*   **Bias pada Hasil Analisis:** Kesimpulan yang keliru akibat data yang salah atau tidak lengkap.
*   **Model Machine Learning Tidak Akurat:** Prediksi yang tidak tepat dari model yang dilatih dengan data "kotor".
*   **Keputusan Bisnis yang Buruk:**  Dampak negatif pada perusahaan akibat keputusan yang didasarkan pada data yang tidak akurat.

**Contoh Data yang Memerlukan Pembersihan:**

Jenis-jenis data yang umum memerlukan pembersihan meliputi:

## **3.1 Mengatasi Missing Value**


**Memahami *Missing Value* dalam Dataset**

***Missing value*** adalah kondisi **ketidaktersediaan atau ketidakcatatan data** yang seharusnya ada dalam dataset.  *Missing value* umum terjadi dan dapat muncul dalam berbagai bentuk karena berbagai alasan, seperti:

*   Data tidak terisi saat pengumpulan.
*   Data hilang selama pemrosesan.

**Jenis-Jenis *Missing Value***

Terdapat beberapa jenis *missing value*, yang diklasifikasikan berdasarkan pola kehilangan datanya:

*   ***Missing Completely at Random (MCAR)***:  Data hilang **sepenuhnya acak** dan **tidak terkait** dengan nilai variabel lain dalam dataset.
*   ***Missing at Random (MAR)***: Data hilang **tidak sepenuhnya acak**, melainkan **berkaitan** dengan **variabel lain** dalam dataset (namun tidak terkait dengan nilai *missing* itu sendiri).
*   ***Missing Not at Random (MNAR)***: Data hilang **tidak acak** dan **berpola**, serta **berkaitan dengan nilai *missing* itu sendiri**.

**Pentingnya Menangani *Missing Value***

Penanganan *missing value* adalah **langkah krusial** dalam *data cleaning*.  *Missing value* dapat berdampak negatif pada:

*   **Hasil Analisis:**  Menyebabkan **bias** dan kesimpulan yang tidak valid.
*   **Model Prediktif:**  Menurunkan **akurasi** dan performa model.

**Teknik Penanganan *Missing Value***

Beberapa teknik umum untuk mengatasi *missing value* meliputi:

*   **Penghapusan (*Deletion*)**:
    *   **Menghapus baris (observasi)** atau **kolom (variabel)** yang memiliki proporsi *missing value* yang tinggi.
    *   **Perlu diperhatikan:** Penghapusan dapat mengurangi ukuran dataset dan menghilangkan informasi berharga jika dilakukan secara berlebihan.
*   **Imputasi (*Imputation*)**:
    *   **Mengisi *missing value* dengan nilai pengganti yang masuk akal**. Beberapa metode imputasi umum:
        *   **Imputasi Sederhana:** Menggunakan **nilai rata-rata (*mean*)**, **nilai tengah (*median*)**, atau **modus (*mode*)** dari kolom.
        *   **Imputasi Lanjutan:** Menggunakan metode yang lebih kompleks, seperti ***k-Nearest Neighbors (KNN)*** atau **regresi (*Regression*)** untuk memprediksi dan mengisi *missing value* berdasarkan variabel lain.

**Kesimpulan:**

Penanganan *missing value* yang tepat adalah **esensial untuk meningkatkan kualitas data** dan **memastikan akurasi analisis serta model prediktif**. Pemilihan teknik penanganan *missing value* perlu disesuaikan dengan jenis *missing value*, karakteristik dataset, dan tujuan analisis.

**Exploratory Data Analysis (EDA): Memahami Data Secara Mendalam**

**Exploratory Data Analysis (EDA)** adalah proses **analisis dataset** untuk:

*   **Merangkum karakteristik utama data.**
*   **Menggunakan visualisasi grafis** sebagai metode utama.

**Tujuan Utama EDA:**

*   **Memahami Data Lebih Baik:**  Memperoleh pengetahuan mendalam tentang isi dan sifat data.
*   **Menemukan Pola:**  Mengungkap tren, hubungan, dan struktur tersembunyi dalam data.
*   **Mendeteksi Anomali:**  Mengidentifikasi nilai-nilai yang tidak biasa atau tidak wajar (*outlier*).
*   **Menguji Hipotesis:**  Memvalidasi atau menolak asumsi awal tentang data.
*   **Memeriksa Asumsi Model:**  Memastikan data memenuhi persyaratan untuk model statistik atau *machine learning* yang akan digunakan.

**Langkah-Langkah Umum dalam EDA:**

1.  **Statistik Deskriptif (*Descriptive Statistics*):**
    *   **Menghitung ukuran ringkasan dasar** untuk memahami **distribusi data**.
    *   Contoh ukuran: **Rata-rata (*Mean*)**, **Median**, **Varians**, **Standar Deviasi**.

2.  **Visualisasi Data (*Data Visualization*):**
    *   **Menggunakan grafik** untuk **memvisualisasikan data**.
    *   Tujuannya: **Melihat distribusi data, hubungan antar variabel, dan *outlier* secara visual.**
    *   Contoh grafik: **Histogram**, **Boxplot**, **Scatter Plot**, **Bar Plot**.

3.  **Pemeriksaan *Outlier* (*Outlier Detection*):**
    *   **Mengidentifikasi data yang tidak sesuai atau ekstrem** yang dapat **mempengaruhi hasil analisis**.
    *   Metode visualisasi (boxplot, scatter plot) dan statistik (Z-score, IQR) dapat digunakan.

4.  **Pemeriksaan Korelasi (*Correlation Analysis*):**
    *   **Menghitung korelasi antar variabel**.
    *   Tujuannya: **Memahami hubungan antar variabel** dalam dataset (positif, negatif, atau tidak berkorelasi).
    *   Metode: **Matriks korelasi, *heatmap*.**

5.  **Pemisahan Data (*Data Segmentation*):**
    *   **Mengelompokkan data** berdasarkan **kategori tertentu.**
    *   Tujuannya: **Memahami distribusi dan pola dalam *subset* data yang berbeda** (misalnya, membandingkan perilaku pembatalan antara jenis hotel yang berbeda).

6.  **Pengujian Hipotesis (*Hypothesis Testing*):**
    *   **Melakukan pengujian statistik** untuk **menguji hipotesis awal** tentang data (misalnya, menguji apakah rata-rata *lead time* untuk pemesanan yang dibatalkan berbeda signifikan dari yang tidak dibatalkan).
    *   Contoh uji: **T-test, ANOVA, Chi-Square test.**

**Manfaat Utama EDA:**

EDA membantu dalam:

*   **Memperoleh *insight* awal** tentang data.
*   **Memandu pengambilan keputusan yang lebih baik dan tepat** dalam tahap analisis selanjutnya.
*   **Memastikan data siap** untuk pemodelan yang lebih kompleks.
*   **Mengurangi risiko kesalahan** dalam analisis.



**Data Preprocessing: Mempersiapkan Data Mentah untuk Analisis**

**Data Preprocessing** adalah **tahap awal yang penting** dalam analisis data dan *machine learning*.  Tujuan utamanya adalah **membersihkan dan mempersiapkan data mentah** agar **siap digunakan dalam model**.

**Apa yang Dilakukan dalam Data Preprocessing?**

*Data preprocessing* melibatkan serangkaian teknik untuk **memperbaiki kualitas data** dan **mengubah formatnya** agar lebih sesuai untuk analisis dan pemodelan. Langkah-langkah utama dalam *data preprocessing* meliputi:

1.  **Pembersihan Data (*Data Cleaning*)**:
    *   **Tujuan:** Mengatasi masalah kualitas data dan membuat data lebih bersih.
    *   **Teknik:**
        *   **Menghapus Data Duplikat:** Menghilangkan baris data yang identik untuk menghindari bias.
        *   **Menangani Data Hilang (*Missing Values*):** Mengisi atau menghapus nilai yang kosong agar data lebih lengkap.
        *   **Mengoreksi Kesalahan Data:** Memperbaiki data yang tidak akurat atau tidak valid (misalnya, kesalahan ketik, nilai di luar rentang yang wajar).

2.  **Transformasi Data (*Data Transformation*)**:
    *   **Tujuan:** Mengubah format atau skala data agar lebih sesuai untuk model atau analisis.
    *   **Teknik:**
        *   **Normalisasi:** Menskalakan data numerik ke rentang tertentu (misalnya 0-1) agar variabel dengan skala berbeda tidak mendominasi.
        *   **Standardisasi:** Menskalakan data numerik agar memiliki rata-rata 0 dan standar deviasi 1, berguna untuk algoritma yang sensitif terhadap skala.
        *   **Mengubah Tipe Data:** Mengkonversi tipe data kolom agar sesuai dengan representasi yang diinginkan (misalnya, mengubah kolom tanggal dari teks menjadi format tanggal).

3.  **Integrasi Data (*Data Integration*)**:
    *   **Tujuan:** Menggabungkan data dari berbagai sumber untuk mendapatkan pandangan yang lebih lengkap.
    *   **Teknik:**
        *   **Menggabungkan Data dari Berbagai Sumber:**  Menyatukan data dari file, database, atau sistem yang berbeda.
        *   **Penggabungan Kolom (*Merging Columns*):**  Menambahkan kolom dari dataset lain berdasarkan kunci yang sama.

4.  **Reduksi Data (*Data Reduction*)**:
    *   **Tujuan:** Mengurangi kompleksitas data tanpa kehilangan informasi penting.
    *   **Teknik:**
        *   **Seleksi Fitur (*Feature Selection*):** Memilih fitur-fitur yang paling relevan dan informatif untuk model dan menghilangkan fitur yang kurang penting atau redundan.
        *   **Reduksi Dimensi dengan PCA (*Principal Component Analysis*):**  Mengurangi jumlah dimensi data dengan mengubah variabel asli menjadi sejumlah kecil komponen utama yang mewakili sebagian besar variasi data.

5.  **Disketisasi Data (*Data Discretization*)**:
    *   **Tujuan:** Mengubah data numerik kontinu menjadi data kategorikal diskrit.
    *   **Teknik:**
        *   ***Binning* (Pengelompokan):** Mengelompokkan nilai numerik ke dalam interval atau *bin* (misalnya, mengelompokkan usia menjadi kelompok usia remaja, dewasa, lansia).
        *   ***One-Hot Encoding*:** Mengubah variabel kategorikal menjadi representasi numerik biner, di mana setiap kategori menjadi kolom baru (misalnya, mengubah kolom "warna" menjadi kolom "warna\_merah", "warna\_biru", "warna\_hijau").

**Mengapa Data Preprocessing Penting?**

Dengan melakukan *data preprocessing*, data menjadi:

*   **Lebih Bersih:** Bebas dari duplikat, *missing value*, dan kesalahan.
*   **Lebih Terstruktur:**  Dalam format yang sesuai untuk analisis dan pemodelan.

Tahap ini **sangat penting** untuk:

*   **Meningkatkan Akurasi Model:** Model yang dilatih dengan data yang telah di-*preprocess* cenderung memberikan prediksi yang lebih tepat.
*   **Meningkatkan Efisiensi Model:** Data yang lebih terstruktur dan tereduksi dapat mempercepat proses pelatihan model dan mengurangi kebutuhan sumber daya komputasi.

### **4.2.1 Overall Outlier Check**

```python
# Fungsi untuk menghitung outlier
def hitung_outlier(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q2 = df[kolom].quantile(0.5)
    Q3 = df[kolom].quantile(0.75)
    batas_atas = Q3 + 1.5 * (Q3 - Q1)
    batas_bawah = Q1 - 1.5 * (Q3 - Q1)
    jumlah_outlier = len(df[(df[kolom] > batas_atas) | (df[kolom] < batas_bawah)])
    persentase_outlier = (jumlah_outlier / len(df)) * 100
    return Q2, batas_bawah, batas_atas, jumlah_outlier, persentase_outlier

# Kolom numerikal
kolom_numerikal = df.describe().columns

# DataFrame untuk menyimpan hasil
tabel_outlier = pd.DataFrame(columns=['Median', 'Batas Bawah', 'Batas Atas', 'Jumlah Outlier', 'Persentase Outlier (%)'])

# Menghitung outlier untuk setiap kolom numerikal
for kolom in kolom_numerikal:
    Q2, batas_bawah, batas_atas, jumlah_outlier, persentase_outlier = hitung_outlier(df, kolom)
    tabel_outlier.loc[kolom] = [Q2, batas_bawah, batas_atas, jumlah_outlier, persentase_outlier]

# Menampilkan hasil
display(tabel_outlier.round(2))
```

Tabel diatas menunjukkan adanya anomali atau outlier pada sebagian besar variabel numerik. Kolom adults dan booking_changes memiliki proporsi outlier yang paling signifikan. Tidak ada outlier pada kolom arrival_date_year, arrival_date_week_number, arrival_date_day_of_month, dan reservation_status_date . ```python

```python
# Function to perform Shapiro-Wilk test and Q-Q plot
def check_normality(column):
    # Shapiro-Wilk test
    stat, p_value = shapiro(df[column])
    
    # Q-Q plot
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    
    plt.subplot(1, 2, 2)
    probplot(df[column], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {column}')
    
    plt.tight_layout()
    plt.show()
    
    return stat, p_value

# Check normality for each column
columns_to_check = ['arrival_date_year', 'arrival_date_week_number', 'arrival_date_day_of_month']
results = {}

for col in columns_to_check:
    stat, p_value = check_normality(col)
    results[col] = {'Shapiro-Wilk Statistic': stat, 'p-value': p_value}

# Display results
results
# Display the conclusion based on Shapiro-Wilk test results
for col, result in results.items():
    stat = result['Shapiro-Wilk Statistic']
    p_value = result['p-value']
    if p_value > 0.05:
        conclusion = 'berdistribusi normal'
    else:
        conclusion = 'tidak berdistribusi normal'
    print(f'Kolom {col} {conclusion} (Shapiro-Wilk Statistic: {stat:.4f}, p-value: {p_value:.4f})')
```

```python
# Pilih kolom yang akan diuji korelasinya
columns_to_correlate = ['lead_time', 'arrival_date_year', 'arrival_date_week_number',
                        'arrival_date_day_of_month', 'stays_in_weekend_nights',
                        'stays_in_week_nights', 'adults', 'children', 'babies',
                        'previous_cancellations', 'previous_bookings_not_canceled',
                        'booking_changes', 'days_in_waiting_list', 'adr',
                        'required_car_parking_spaces', 'total_of_special_requests',
                        'reservation_status_date', 'arrival_date', 'visitor']

# Hitung matriks korelasi Spearman
correlation_matrix = df[columns_to_correlate].corr(method='spearman')

# Buat heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Spearman Correlation Heatmap')
plt.show()

```

**Penjelasan Heatmap Korelasi Spearman**

Heatmap yang disajikan memberikan visualisasi yang komprehensif mengenai hubungan antar variabel numerik dalam dataset, dengan menggunakan koefisien korelasi Spearman. Korelasi Spearman mengukur kekuatan dan arah hubungan monoton antara dua variabel tanpa mengasumsikan distribusi data yang normal.

Kriteria uji korelasi biasanya dikategorikan sebagai berikut:

- **Tingkat Korelasi Rendah**: Nilai korelasi antara 0.00 hingga ±0.30
- **Tingkat Korelasi Sedang**: Nilai korelasi antara ±0.30 hingga ±0.70
- **Tingkat Korelasi Tinggi**: Nilai korelasi antara ±0.70 hingga ±1.00

**Interpretasi Warna dan Nilai Korelasi:**

* **Warna:**
    * Merah Tua: Korelasi positif yang kuat (semakin besar nilai positif, semakin searah hubungan antara dua variabel).
    * Biru Tua: Korelasi negatif yang kuat (semakin besar nilai negatif, semakin berlawanan arah hubungan antara dua variabel).
    * Warna Netral (Putih atau Kuning): Korelasi lemah atau tidak ada korelasi.
* **Nilai Korelasi:**
    * -1: Korelasi negatif sempurna (ketika satu variabel meningkat, variabel lainnya menurun).
    * 0: Tidak ada korelasi.
    * +1: Korelasi positif sempurna (kedua variabel bergerak searah).

1. **No Deposit:** Jenis transaksi di mana pelanggan tidak diwajibkan untuk melakukan pembayaran di muka. Pembayaran penuh biasanya dilakukan saat check-in atau setelah menikmati layanan. Opsi ini memberikan fleksibilitas bagi pelanggan, namun juga dapat meningkatkan risiko pembatalan.

* **Refundable Deposit:** Menawarkan keseimbangan antara fleksibilitas bagi pelanggan dan perlindungan bagi penyedia layanan.
